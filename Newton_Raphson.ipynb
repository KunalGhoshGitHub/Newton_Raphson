```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from IPython import display
from moviepy.editor import *
from sympy import *

%matplotlib notebook

fig, ax = plt.subplots(figsize=(10,5.625))

k = 1.2

# Set the range for the axis
x_limit = 2
y_limit = x_limit*9/16
ax.set_xlim(-x_limit,x_limit)
ax.set_ylim(-y_limit,y_limit)

class Func:
    
    def f(self,x):
        y = 2*x**2
        return y
    
    def diff_f(self,x):
        y = diff(self.f(x),x)
        return y
    
    def integrate_f(self,x):
        y = integrate(self.f(x),x)
        return y
            
# Creating a emp instance of Employee class    
x = Symbol('x')
f1 = Func()
f = f1.f(x) 
np_f = lambdify(x,f)
diff_f = f1.diff_f(x)
np_diff_f = lambdify(x,diff_f)
       
files = []

x_data = np.linspace(k,-k,100)
y_data = np_f(x_data)

def newton_raphson_all_iteration(x_0,n):
    p = np.array(x_0)
    for i in range(0,n,1):
        m = x_0 - (np_f(x_0)/np_diff_f(x_0))
        # print(m)
        p = np.append(p,m)
        x_0 = m
        # print(p.shape)
    return p

n = 10
x_0 = 1
x_iter = newton_raphson_all_iteration(x_0,n)
y_iter = np_f(x_iter)

def animation_frame(i):
    # The x axis and y axis
    # x axis
    plt.plot([-x_limit,x_limit],[0,0],color = "#d9d9d9")
    # y axis
    plt.plot([0,0],[-y_limit,y_limit],color = "#d9d9d9")
    n = 10
    x_0 = 0.8
    x_iter = newton_raphson_all_iteration(x_0,n)
    y_iter = np_f(x_iter)
    plt.plot(x_iter,y_iter,"go")
    plt.plot(x_iter,y_iter*0,"mo")
    for j in range(0,n,1):
        plt.plot([x_iter[j],x_iter[j]],[y_iter[j],0],linewidth = 1)
        plt.plot([x_iter[j],x_iter[j+1]],[y_iter[j],0],linewidth = 1)
    # This is the graph to be displayed
    plt.plot(x_data,y_data,"g")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    # plt.legend(["","","b = "+str(f"{i:.2f}"),"b = 0"])
    plt.title("Iteration: 1")
    plt.xticks([])
    plt.yticks([])

anim = FuncAnimation(fig, func=animation_frame, frames=range(0, 100, 1), interval=2)
plt.show()
writervideo = animation.FFMpegWriter(fps=50) #fps = Frame per second (This is the speed of the  animation) 
anim.save("Iteration_1.mp4", writer=writervideo, dpi = 250) #dpi = dots per inch (This is the quality of the image)
```
